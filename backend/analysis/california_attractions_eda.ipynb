```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r'C:\Users\turne\OneDrive\Desktop\TripleTen Project Files\Code Jam Projects\california_attractions_data.csv')
```

```python
top_cities_df = df['city'].value_counts().head(10).reset_index()
sns.barplot(data = top_cities_df, 
               y = 'city',
               x = 'count', 
               color = 'green')
plt.title('Top 10 Cities by Number of Attractions in California')
plt.xlabel('Number of Attractions')
plt.ylabel('City')
plt.show()
```

This graph shows the top 10 cities in California based on number of attractions from our dataset. Los Angeles taking the top spot makes sense as even though our dataset isn't all encompassing, it is still the largest city in California and would tend to have the highest number of sights to see in and around the city. 

```python
top_attractions_df = df['category'].value_counts().head(11).reset_index()
top_attractions_df = top_attractions_df[top_attractions_df['category'] != 'Other'] #Other category is not useful for analysis, it has nearly twice as many entries as the next highest category without a clear grouping
sns.barplot(data = top_attractions_df, 
               y = 'category',
               x = 'count', 
               color = 'blue')
plt.title('Top 10 Categories of Attractions in California')
plt.xlabel('Count')
plt.ylabel('Category')
plt.show()
```

A second graph that makes sense in the context of our project. California is the center of the US's film and television industries, so logic would follow they would have a lot of movie theaters, which are present in our dataset. There are also plenty of professional, amateur, and college sports teams in the state, and that's not even accounting for local sports complexes or sports themed attractions, all of which are popular within the state.

```python
#Splitting dataset into Northern, Central, and Southern California to determine what attractions are most popular in each region
def region_split(latitude):
    if latitude >= 38.5:
        return 'Northern California'
    elif latitude < 38.5 and latitude >= 36:
        return 'Central California'
    else:
        return 'Southern California'
df['region'] = df['latitude'].apply(region_split)

region_df = df['region'].value_counts()
```

```python
sns.barplot(data = region_df.reset_index(), 
               y = 'region',
               x = 'count', 
               palette = 'Blues_d')
plt.title('Number of Attractions by Region in California')
plt.xlabel('Number of Attractions')
plt.ylabel('Region')
plt.show()
```

Looking at this graph, we see that most of our attractions are skewed towards Southern California. With the context of the geography of the state, it's easy to understand that most of the population lives in that region and would therefore needs things to do and sights to see!

```python
socal_df = df[df['region'] == 'Southern California'].reset_index(drop = True)
norcal_df = df[df['region'] == 'Northern California'].reset_index(drop = True)
centralcal_df = df[df['region'] == 'Central California'].reset_index(drop = True)
```

```python
top_attractions_socal_df = socal_df['category'].value_counts().head(11).reset_index()
top_attractions_socal_df = top_attractions_socal_df[top_attractions_socal_df['category'] != 'Other'] #Other category is not useful for analysis, it has nearly twice as many entries as the next highest category without a clear grouping
sns.barplot(data = top_attractions_socal_df, 
               y = 'category',
               x = 'count', 
               palette = 'pastel')
plt.title('Top 10 Categories of Attractions in Southern California')
plt.xlabel('Number of Attractions')
plt.ylabel('City')
plt.show()
```

The top 3 spots in Southern California based on number of attractions follows the same logic. Hollywood, 4 professional sports teams, and an immense number of hotels can be found in just Los Angeles alone, and when you add in the rest of the region those numbers increase even more!

```python
top_attractions_norcal_df = norcal_df['category'].value_counts().head(11).reset_index()
top_attractions_norcal_df = top_attractions_norcal_df[top_attractions_norcal_df['category'] != 'Other'] #Other category is not useful for analysis, it has nearly twice as many entries as the next highest category without a clear grouping
sns.barplot(data = top_attractions_norcal_df, 
               y = 'category',
               x = 'count', 
               palette = 'pastel')
plt.title('Top 10 Categories of Attractions in Northern California')
plt.xlabel('Number of Attractions')
plt.ylabel('City')
plt.show()
```

Historical Buildings take the top spot in Northern California, primarily due to the strong historical preservation efforts in San Francisco, the region's most populous city. Most of these buildings were built during and shortly after the California Gold Rush, and the influx of wealth led to the creation of many more.

```python
top_attractions_centralcal_df = centralcal_df['category'].value_counts().head(11).reset_index()
top_attractions_centralcal_df = top_attractions_centralcal_df[top_attractions_centralcal_df['category'] != 'Other'] #Other category is not useful for analysis, it has nearly twice as many entries as the next highest category without a clear grouping
sns.barplot(data = top_attractions_centralcal_df, 
               y = 'category',
               x = 'count', 
               palette = 'pastel')
plt.title('Top 10 Categories of Attractions in Central California')
plt.xlabel('Number of Attractions')
plt.ylabel('City')
plt.show()
```

Central California is a smaller subset of the state, but they certainly follow in the footsteps of the their neighbors by having a large amount of theaters and venues to enjoy the entertainment created down in Hollywood. They also have a fair share of professional sports teams and venues to visit on your travels as well!
